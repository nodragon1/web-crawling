# IMBD 평점이 가장 높은 영화

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating'
res = requests.get(url)

soup = BeautifulSoup(res.text, 'html.parser')

movies = soup.find_all('div', class_='lister-item mode-advanced')

titles = []
releases = []
ratings = []

for movie in movies:
    title = movie.h3.a.text
    titles.append(title)
    
    release = movie.h3.find('span').findNext('span').text
    releases.append(release)
    
    rating = movie.find('strong').text
    ratings.append(rating)
    
df = pd.DataFrame({'Title': titles,
                   'Release_date': releases,
                   'Rating': ratings})

df.head(20)
```

# 특정 연도의 프로야구 순위표

```python
while True:
    year = int(input('연도를 선택하세요.: '))
    if 1991 <= year <= 2022:
        break

url = f'https://sports.news.naver.com/kbaseball/record/index?category=kbo&year={year}'
res = requests.get(url)

soup = BeautifulSoup(res.text, 'html.parser')

table = soup.find('tbody', attrs={'id': 'regularTeamRecordList_table'}).find_all('tr')

순위 = []
팀명 = []
경기수 = []
승 = []
패 = []
무 = []
승률 = []
게임차 = []

for t in table:
    rank = t.strong
    순위.append(rank.text)
    
    team = t.find('span').findNext('span')
    팀명.append(team.text)
    
    games = team.findNext('span')
    경기수.append(games.text)
    
    win = games.findNext('span')
    승.append(win.text)
    
    lose = win.findNext('span')
    패.append(lose.text)
    
    draw = lose.findNext('span')
    무.append(draw.text)
    
    rate = draw.findNext('strong')
    승률.append(rate.text)
    
    diff = rate.findNext('span')
    게임차.append(diff.text)

df = pd.DataFrame({'순위': 순위,
                   '팀명': 팀명,
                   '경기수': 경기수,
                   '승': 승,
                   '패': 패,
                   '무': 무,
                   '승률': 승률,
                   '게임차': 게임차})

df
```
